```python
def add_values_to_away_player(normalized_df):
    '''
    If the metrics have less than 11 columns (players) are add
    0 to complet the values (avoid the image with less pixels).
    '''
    if len(normalized_df.columns) < 11:
        normalized_df[10] = 0
    return normalized_df
```

```python
#Lê os arquivos de métricas de grafos 
def read_dataframe_metric(file):
    #metric = pd.read_csv('grafo//'+ partida+ '/'+graph_metrics[metric_count]+'.csv')
    metric = pd.read_csv(mypath+'/'+str(file))
    metric.drop(metric.columns[[0]],1, inplace= True)
    
    #Normalize the data (entre 1 and 0) to avoid error in Imageio
    metric_normalized = (metric-metric.min())/(metric.max()-metric.min())
    
    #Fix to match without 10 players in field
    metric_normalized = add_values_to_away_player(metric_normalized)
    
    metric_normalized.fillna(0,inplace= True)
    
    return metric_normalized
```

```python
def read_files_on_fold():
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    return onlyfiles
```
